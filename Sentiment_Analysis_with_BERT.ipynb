# Data Preparation for Bert Model

```python
dataset_id = 'imdb'
dataset = load_dataset(dataset_id)
```

```python
dataset
```

```python
dataset['train'][0]
```

## Bert Model

```python
model_id = 'bert-base-uncased'
tokenizer = BertTokenizerFast.from_pretrained(model_id)
```

```python
def preprocess_function(examples):
    return tokenizer(examples['text'], truncation=True, padding=True)
```

```python
tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
tokenized_dataset
```

```python
tf_train_dataset = tokenized_dataset["train"].to_tf_dataset(
    columns=['input_ids', 'token_type_ids', 'attention_mask', 'label'],
    shuffle=True,
    batch_size=BATCH_SIZE,
)
```

```python
tf_train_dataset
```

```python
tf_val_dataset = tokenized_dataset["test"].to_tf_dataset(
    columns=['input_ids', 'token_type_ids', 'attention_mask', 'label'],
    shuffle=True,
    batch_size=BATCH_SIZE,
)
```

```python
def swap_positions(dataset):
  return {'input_ids':dataset['input_ids'], 'token_type_ids':dataset['token_type_ids'],'attention_mask':dataset['attention_mask'],} ,dataset['label']
```

```python
tf_train_dataset = tf_train_dataset.map(swap_positions).prefetch(tf.data.AUTOTUNE)
tf_val_dataset   = tf_val_dataset.map(swap_positions).prefetch(tf.data.AUTOTUNE)
```

# Training

```python
num_epochs = 3
batches_per_epoch = len(tokenized_dataset['train']) // BATCH_SIZE
total_train_steps = int(batches_per_epoch * num_epochs)
```

```python
optimizer, schedule = create_optimizer(init_lr=2e-5, num_warmup_steps=0, num_train_steps=total_train_steps)
```

```python
model.compile(optimizer=optimizer, metrics=['accuracy'])
```

```python
history = model.fit(tf_train_dataset, validation_data=tf_val_dataset, epochs=num_epochs)
```
